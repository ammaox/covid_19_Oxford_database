```python
import numpy as np
import pandas as pd
```

```python
regions_list = pd.read_csv("out/IVS_grouped_by_country_region.csv", encoding = 'utf-16', low_memory = False)
regions_list = regions_list[['Survey', 'Wave', 'Country Numeric', 'Country ISO3', 'Region Numeric',
       'Region Name', 'NUTS1', 'NUTS2', 'NUTS3', 'Sample Size']]
```

```python
most_recent = pd.DataFrame(columns=regions_list.columns)


for idx in regions_list["Country Numeric"].unique():
    mask = regions_list["Country Numeric"] == idx
    
    name = regions_list[ mask ]["Country ISO3"].unique()[0]
    
    EVS_waves = regions_list[ mask & (regions_list["Survey"] == "EVS") ]["Wave"].unique()
    
    
    if len(EVS_waves)>0:
        most_recent = most_recent.append(regions_list.loc[mask & (regions_list["Survey"] == "EVS")
                                        & (regions_list["Wave"] == EVS_waves[-1]),:])
    
    WVS_waves = regions_list[ mask & (regions_list["Survey"] == "WVS") ]["Wave"].unique()
    
    if len(WVS_waves)>0:
        most_recent = most_recent.append(regions_list.loc[mask & (regions_list["Survey"] == "WVS")
                                        & (regions_list["Wave"] == WVS_waves[-1]),:])
```
