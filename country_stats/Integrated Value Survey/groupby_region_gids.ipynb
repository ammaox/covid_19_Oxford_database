```python
import numpy as np
import pandas as pd
```

```python
region_gid = pd.read_csv("gid/region_codes_dictionary.csv")
```

```python
region_array_gid = region_gid.groupby(list(region_gid.columns[:-1])).agg({'gid':lambda x: ':'.join(list(x))})
```

```python
region_array_gid = region_array_gid.reset_index()
```

```python
region_array_gid
```

```python
region_array_gid.to_csv("gid/region_array_gid.csv", index=False)
```
